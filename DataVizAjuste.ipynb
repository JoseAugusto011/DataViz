```python
import pandas as pd
from matplotlib.pyplot import subplots
import seaborn as sns

# pandas
df = pd.read_csv('preco-combs-pb-2022-02.csv')


```

```python
# matplotlib
fig, ax = subplots(figsize=(14,5))
ax.set_title('Preço de combustíveis - PB - Semestre 2022/01. Fonte: ANP')
ax.set_ylabel('litro (R$)')

# seaborn
sns.set_style("whitegrid")
g = sns.boxplot(data=df, x="Bandeira", y="Valor de Venda",
    hue='Produto',palette="deep",ax = ax)
sns.move_legend(g, "upper center", title='', ncol=1,
    frameon=True, bbox_to_anchor=(1.11, .8))
sns.despine(offset=5,trim=True)
sns.set_style()
```

```python
listaCombustiveis = list(set(df['Produto'].values))

for nome in listaCombustiveis:


  # matplotlib
  fig, ax = subplots(figsize=(14,5))
  ax.set_title(f'Preço  {nome} - PB - Semestre 2022/01. Fonte: ANP')
  ax.set_ylabel('litro (R$)')

  # seaborn
  sns.set_style("whitegrid")
  g = sns.boxplot(data=df[df['Produto']==f'{nome}'], x="Bandeira", y="Valor de Venda",hue='Produto',palette="deep",ax = ax)
  ax.set(ylim=(4.5,8.5))
  sns.move_legend(g, "upper center", title='', ncol=1,frameon=True, bbox_to_anchor=(1.11, .8))
  sns.despine(offset=5,trim=True)

  sns.set_style()
```
